```python
# Keep only 'Country Name' and recent year columns
df = df[['Country Name', '2022']]  # You can change '2022' to any year that has more data
df = df.dropna()  # Drop rows with missing values
df.head()
```

```python
# Drop NaN population values
df = df[df['2022'].notna()]

# Sort by population in descending order
df_sorted = df.sort_values(by='2022', ascending=False)

# Show top 10 countries
df_sorted.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

top_10 = df_sorted.head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='2022', y='Country Name', data=top_10, palette='Set2')
plt.title('Top 10 Most Populated Countries in 2022')
plt.xlabel('Population')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```
