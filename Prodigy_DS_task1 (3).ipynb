```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your file (your version v2_71683)
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_71683.csv", skiprows=4)

# Keep only relevant columns
df = df[['Country Name', '2022']]

# Remove grouped rows that are not real countries
remove_list = ['World', 'IDA', 'IBRD', 'income', 'demographic', 'total', 'East Asia']
df = df[~df['Country Name'].str.contains('|'.join(remove_list), case=False)]

# Remove NaN values
df = df[df['2022'].notna()]

# Sort by 2022 population
df_sorted = df.sort_values(by='2022', ascending=False).head(10)

# Plot bar chart
plt.figure(figsize=(12,6))
sns.barplot(data=df_sorted, x='Country Name', y='2022', palette='coolwarm')
plt.xticks(rotation=45)
plt.title("Top 10 Most Populated Countries in 2022", fontsize=14)
plt.xlabel("Country")
plt.ylabel("Population")
plt.tight_layout()
plt.show()
```
